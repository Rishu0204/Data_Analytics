```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"D:\Data_Analyst\sample_files\world_populationEDA.csv")
df
```

```python
#to the change the decimal format of the data
pd.set_option('display.float_format',lambda x: '%.2f' % x)
```

```python
df.info()
```

```python
df.describe()
```

```python
#to see how many null values we have in a df
df.isnull().sum()
```

```python
#to see the number of unique records in each column
df.nunique()
```

```python
#to sort the dataframe
#the sort_value() by default sort in acscending order
#to sort in descending order we can specify ascending parameter as False which is by default true
df.sort_values(by='2022 Population').head()      #head() just show the top 5 values. It can also be specified to show any number of data
```

```python
#seeing correlation between various columns
#the corr() compare every column with every column to see how correlated they are
#mentioning numeric_only parameter is important is neccessary as under new update the numeric_only is set to False by default
#this may cause conflict if the dataframe consist of string values
df.corr(numeric_only=True)
```

```python
#visualizing correlation
#we will use heatmap to visualize 
#heatmap gives a colour code to each relation like highly correlated columns will have similar color unlike the less correlated columns
sns.heatmap(df.corr(numeric_only=True),annot=True)
plt.rcParams['figure.figsize']=(20,10)
plt.show()
```

```python
#grouping the data to see the trends
#in this case of world population we will groupby the continents to see the trend of growth using mean()
df.groupby('Continent').mean('2022 Population').sort_values('2022 Population',ascending=False)
```

The above query give the insight that the asia has the highest avg. population also they have the least rank average that means most of the countries of asia ranks higher i.e they have lesser ranks, explaining lower rank average. Similarly for other continents as well.

```python
#checking the contents of countries of the continent Oceania to gain more insight
df[df['Continent'].str.contains('Oceania')].sort_values('Rank')
```

```python
#visualizing the population trends
# df2=df.groupby('Continent').mean('2022 Population').sort_values('2022 Population',ascending=False)

#updating df2 to our requirement i.e to have only population columns 

df2=df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean('2022 Population').sort_values('2022 Population',ascending=False)
df2
```

```python
#to find the index that we want
df.columns
```

```python
#now if we try to plot this df2 we will not be able to understand anything from it
df2.plot()
```

```python
#this is because it is visualizing by continent not the population
#to do that first we need to transpose the df2
df3=df2.transpose()
df3
```

```python
#this plot is lot better but till not right as it is visualizing columns like rank,area,growth,etc. which we don't want
#this would mean that we have to correct our df2 frame to only have population columns
#after updating if we plot again we will get much better visualization of the population trend
df3.plot()
```

```python
#finding outlier
#we use box plot to find outliers
df3.boxplot()
#as we can see we have one outlier in the europe continent
```
