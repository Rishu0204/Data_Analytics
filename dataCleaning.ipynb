DATA CLEANING IN PANDAS


```python
import pandas as pd
```

```python
df=pd.read_excel(r"D:\Data_Analyst\sample_files\Customer Call List.xlsx")
df
```

```python
#removing duplicates
df=df.drop_duplicates()
df
```

```python
#removing not useful columns
df=df.drop(columns='Not_Useful_Column')
df
```

```python
#use of strip() to remove the errors in a string
#eg: in last name column ...Potter has '...' in the beginning which is not desired 
#to remove it we can use strip()
#the strip() remove the unwanted strings from the start and beginning of each mentioned record

# df['Last_Name']=df['Last_Name'].str.strip('...')

#using this method we need to specify each strip individully for each error
#better method to use the strip is to name everything in one strip 
df['Last_Name']=df['Last_Name'].str.strip('123._/')
df
#the lstrip only strips from left side of string
# the rstring only strip from right side of string 
```

```python
#fixing the format of the phone number
#we will use the replace function to do that
#first we will remove any unwanted symbol or character in between phone number
#eg: '876|678|3469' here '|' is a unwanted character

df['Phone_Number']=df['Phone_Number'].str.replace('[^a-zA-Z0-9]','',regex=True)
```

```python
#now formatting the phone number
#we can use for loop or lambda function to do that

# df['Phone_Number']=df['Phone_Number'].apply(lambda x: str(x))

#now adding '-' after every 3 number

df['Phone_Number']=df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
#now removing thr nan and na
df['Phone_Number']=df['Phone_Number'].str.replace('nan--','')
df['Phone_Number']=df['Phone_Number'].str.replace('Na--','')
df
```

```python
#now we need to split the address column into 3 other columns i.e. street,state and zip code
#as we can see they are seperated by a comma, so we can use split() to acheive this

df[['Street_Address','State','Zip_Code']]=df['Address'].str.split(',',expand=True)
df
```

```python
#formatting paying customer and do not contact
df['Paying Customer']=df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer']=df['Paying Customer'].str.replace('No','N')
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('No','N')
df
```

```python
#removing all nan and n/a values
df=df.fillna('')
df=df.replace('N/a','')
df
```

```python
#now we need to see wether the customer wants to be called or not
#i.e. need to to find 'Y' in 'Do_Not_Contact' column, if found we need to drop that record

for x in df.index:
    if df.loc[x,"Do_Not_Contact"]=='Y':
        df.drop(x,inplace=True)
df
```

```python
#now we also need to remove customer that doesn't have a phone number
for x in df.index:
    if df.loc[x,"Phone_Number"]=='':
        df.drop(x,inplace=True)
df
```
