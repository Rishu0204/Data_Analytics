PANDAS VISUALIZATION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"D:\Data_Analyst\sample_files\populationVisulization.csv")
df=df.set_index('City')
df
```

```python
#list all the styles available in matplotlib
print(plt.style.available)
#we can use these styles to customize our charts
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
#by default it uses line graph
#we can use kind to also specify what kind of plot we want
df.plot(kind='line')
```

```python
#to seperate the graph we use subplots parameter and set it as true
df.plot(kind='line',subplots= True)
```

```python
#we can also give title to the graph and label the axis
df.plot(kind='line',title='City Population',xlabel='City',ylabel='Population')
```

```python
#bar chart
df.plot(kind='bar')
```

```python
#we can also stack the bar graph
df.plot(kind='bar',stacked=True)
```

```python
# we can also specify the graph to a single or multiple columns instead of all the columns
df['2022 Population'].plot(kind='bar')
```

```python
#horizontal bar chart
df.plot.barh()
```

```python
#scatter chart
df.plot.scatter(x='2022 Population',y='2020 Population')
```

```python
# we can also control the size and color of the dots
df.plot.scatter(x='2022 Population',y='2020 Population',s=50,c='Red')
```

```python
#histogram chart
#bins means the no. of bars on a histogram
df.plot.hist(bins=50)
```

```python
#box plot
df.plot.box()
```

```python
#area chart
#we can adjust the size using figsize
df.plot.area(figsize=(10,2))
```

```python
#pie chart
df.plot.pie(y='2022 Population',figsize=(10,10))
```
