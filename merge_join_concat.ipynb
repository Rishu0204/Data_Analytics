```python
import pandas as pd
```

```python
df1=pd.read_csv(r"D:\Data_Analyst\sample_files\sample1.csv")
df1
```

```python
df2=pd.read_csv(r"D:\Data_Analyst\sample_files\sample2.csv")
df2
```

```python
#merge
#works similar to inner join
df1.merge(df2)
```

```python
#inner join using merge
#inner join gives only whats common in both left and right tables
#how specify what join we are performing
#on specify on which column we are performing join
df1.merge(df2,how='inner',on='ID')
```

```python
#outer join using merge
#outer join give everything regardless wethear it is unique in both tables or not
#outer join sees the left table see whats in common in both prints it then print whats not common
#like how ID 1001 and 1002 are common in both so all their details are filled 
#1003 and 1004 were not common in both so they have NaN in their age column which is in the right table
df1.merge(df2,how='outer')
```

```python
#left join using merge
#left join gives everything in left table and also the details which are common to both left and right
df1.merge(df2,how='left')
```

```python
#right join using merge
#right join gives everything in right table and also the details which are common to both left and right
df1.merge(df2,how='right')
```

```python
#cross join
#map each element of left table with the right table
df1.merge(df2,how='cross')
```

```python
#join()
df1.join(df2,on='ID',how='outer',lsuffix='_Left',rsuffix='_Right')
```

```python
#above is not the right method to use the join()
#join() doesn't work well with columns
#we use join() with indexs
df3=df1.set_index('ID').join(df2.set_index('ID'),lsuffix='_Left',rsuffix='_Right')
df3
```

```python
#concat()
#concat will put the table on upon another
pd.concat([df1,df2])
```
