```python
 import pandas as pd
```

```python
df=pd.read_csv(r"D:\Data_Analyst\sample_files\flavors.csv")
df
```

```python
#group by function 
df.groupby('Base Flavor')
```

```python
group_by_frame=df.groupby('Base Flavor')
```

```python
#mean()
group_by_frame.mean('Flavor Rating')
```

```python
#direct method to use agg functions
df.groupby('Base Flavor').mean('Flavor Rating')
```

```python
#count()
df.groupby('Base Flavor').count()
```

```python
#sum
df.groupby('Base Flavor').sum('Flavor Rating')
```

```python
#min()
df.groupby('Base Flavor').min()
```

```python
#max
df.groupby('Base Flavor').max()
```

```python
#agg function
df.groupby('Base Flavor').agg({'Flavor Rating':['mean','min','max','sum']})
```

```python
#to use agg() on multiple columns
df.groupby('Base Flavor').agg({'Flavor Rating':['mean','min','max','sum'],'Texture Rating':['mean','min','max','sum']})
```

```python
#grouping by multiple columns
df.groupby(['Base Flavor','Liked']).mean('Flavor Rating')
```

```python
#describe()
df.groupby('Base Flavor').describe()
```
